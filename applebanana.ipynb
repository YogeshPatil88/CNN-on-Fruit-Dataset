```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
import os
#import pickle
```

```python
DataDir= r'C:\Users\shree\Downloads\appleorbanana\train'

CATEGORIES=['Apple Braeburn','Banana']
for i in CATEGORIES:
    path=os.path.join(DataDir,i)
    for img in os.listdir(path):
        img_array=cv2.imread(os.path.join(path,img),cv2.IMREAD_GRAYSCALE)
        plt.imshow(img_array,cmap='gray')
        plt.show()
        break
    break
```

```python

img_size=100

new_array=cv2.resize(img_array,(img_size,img_size))
plt.imshow(new_array,cmap='gray')
plt.show()
```

```python
training_data=[]

def create_training_data():
    for i in CATEGORIES:

        path=os.path.join(DataDir,i)
        class_num=CATEGORIES.index(i)

        for img in os.listdir(path):
            try:
                img_array=cv2.imread(os.path.join(path,img),cv2.IMREAD_GRAYSCALE)
                new_array=cv2.resize(img_array,(img_size,img_size))
                training_data.append([new_array,class_num])
            
            except Exception as e:
                pass
```

```python
create_training_data()
print(len(training_data))
            
import random
random.shuffle(training_data)
for sample in training_data[:10]:
    print(sample)
```

```python
X=[]
y=[]

for features,label in training_data:
    X.append(features)
    y.append(label)

print(X[0].reshape(-1,img_size,img_size,1))
```

```python
X=np.array(X).reshape(-1,img_size,img_size,1)
```

```python
import tensorflow as tf
from keras.models import Sequential
from keras.layers import Dense,Dropout,Activation,Flatten
import pickle

from keras.layers import Conv2D,MaxPooling2D

#pickle_in=open(r"C:\Python37\Projects\Youtube-Tutorial-Live\X.pickle","rb")
#X=pickle.load(pickle_in)

#pickle_in=open(r"C:\Python37\Projects\Youtube-Tutorial-Live\y.pickle","rb")
#y=pickle.load(pickle_in)

print(X)
```

```python
X=X/255.0
```

```python
X
```

```python
model= Sequential()

model.add(Conv2D(64,(3,3),input_shape=X.shape[1:]))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2,2)))

model.add(Conv2D(64,(3,3)))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Flatten())

model.add(Dense(128, input_shape = X.shape[1:], activation = 'relu'))

model.add(Dense(2, activation = 'sigmoid'))
```

```python
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
model.fit(X,y,batch_size=4,epochs=5,validation_split=0.2)
```

```python
image=r'C:\Users\shree\Downloads\appleorbanana\test\0012.jpg'


def prepare(image):
    img_size=100
    img_array=cv2.imread(image,cv2.IMREAD_GRAYSCALE)
    new_array=cv2.resize(img_array,(img_size,img_size))
    return new_array.reshape(-1,img_size,img_size,1)
```

```python
prediction=model.predict([prepare(image)])
print(CATEGORIES[int(prediction[0][1])])
```
